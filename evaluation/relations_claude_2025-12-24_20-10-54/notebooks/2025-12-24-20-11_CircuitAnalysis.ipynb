```python
# Read the original documentation (PDF)
# First, let's try to extract text from the PDF
import subprocess

original_doc_pdf = '/net/scratch2/smallyan/relations_eval/documentation.pdf'

# Try using pdftotext if available
try:
    result = subprocess.run(['pdftotext', original_doc_pdf, '-'], capture_output=True, text=True)
    if result.returncode == 0:
        original_doc_text = result.stdout
        print("=== Original Documentation (PDF) ===")
        print(original_doc_text[:5000])
        print("\n... [truncated for display]")
    else:
        print("pdftotext failed, trying alternative method")
except FileNotFoundError:
    print("pdftotext not available, trying PyPDF2")
```

```python
# Get the full original documentation text
print("=== Full Original Documentation ===")
print(f"Total characters: {len(original_doc_text)}")
print("\n" + original_doc_text)
```

```python
# Let's save the original doc text in parts and read it more carefully
# First, let's extract the key sections

# Split by sections
sections = original_doc_text.split('\n\n')
print(f"Number of sections: {len(sections)}")

# Find abstract
for i, section in enumerate(sections[:20]):
    if 'ABSTRACT' in section or 'Abstract' in section:
        print(f"\n=== ABSTRACT (section {i}) ===")
        print(section[:2000])
        break
```

```python
# Let's extract key parts of the original documentation more systematically
import re

# Find the abstract
abstract_match = re.search(r'A BSTRACT\s*(.*?)(?=\n\d+\s+I NTRODUCTION|\n1\s+I NTRODUCTION)', original_doc_text, re.DOTALL)
if abstract_match:
    abstract = abstract_match.group(1).strip()
    print("=== ABSTRACT ===")
    print(abstract[:2000])
```

```python
# Extract key results sections from the original documentation
# Looking for experimental results, metrics, and conclusions

# Find conclusion section
conclusion_match = re.search(r'(?:CONCLUSION|C ONCLUSION|6\s+C ONCLUSION)(.*?)(?=R EFERENCES|REFERENCES|A\s+A PPENDIX)', original_doc_text, re.DOTALL | re.IGNORECASE)
if conclusion_match:
    conclusion = conclusion_match.group(1).strip()
    print("=== CONCLUSION ===")
    print(conclusion[:3000])
else:
    print("Conclusion section not found with regex, searching differently...")
    # Try to find it by looking for "Conclusion" keyword
    if "ONCLUSION" in original_doc_text:
        idx = original_doc_text.find("ONCLUSION")
        print(original_doc_text[idx-2:idx+3000])
```

```python
# Extract key experimental results - looking for sections 4 and 5
# Section 4: Experiments on accuracy, faithfulness etc.

# Find Section 4
section4_match = re.search(r'4\s+E XPERIMENTS?(.*?)(?=5\s+)', original_doc_text, re.DOTALL)
if section4_match:
    section4 = section4_match.group(1).strip()
    print("=== SECTION 4: EXPERIMENTS ===")
    print(section4[:4000])
else:
    print("Section 4 not found")
```

```python
# Continue reading section 4 and section 5
section5_match = re.search(r'5\s+A PPLICATIONS?(.*?)(?=6\s+C ONCLUSION)', original_doc_text, re.DOTALL)
if section5_match:
    section5 = section5_match.group(1).strip()
    print("=== SECTION 5: APPLICATIONS ===")
    print(section5[:4000])
else:
    print("Section 5 not found")
```

```python
# Look for section 5 with different patterns
print("Searching for section 5...")
# Try different patterns
patterns = [
    r'5\s+A\s*PPLICATIONS',
    r'5\s+APPLICATIONS', 
    r'5\s+A TTRIBUTE',
    r'(?<=\n)5\s+[A-Z]'
]

for pattern in patterns:
    match = re.search(pattern, original_doc_text)
    if match:
        print(f"Found with pattern '{pattern}' at position {match.start()}")
        print(original_doc_text[match.start():match.start()+3000])
        break
else:
    # Just find what's between section 4 results and conclusion
    print("Looking for content after section 4...")
    idx = original_doc_text.find("4.2")
    if idx > 0:
        print(original_doc_text[idx:idx+4000])
```

```python
# Let's find the key results more systematically - look for figures and tables
print("=== Key Figures and Tables in Original Doc ===")
figures = re.findall(r'Figure \d+[^.]*\.', original_doc_text)
for fig in figures[:15]:
    print(f"- {fig}")

print("\n=== Tables ===")
tables = re.findall(r'Table \d+[^.]*\.', original_doc_text)
for tbl in tables[:10]:
    print(f"- {tbl}")
```

```python
# Extract key quantitative results mentioned in the paper
print("=== Key Quantitative Results from Original Documentation ===\n")

# Search for percentage mentions and key metrics
percent_mentions = re.findall(r'\d+(?:\.\d+)?%[^.]*\.', original_doc_text)
print("Percentage mentions:")
for p in percent_mentions[:15]:
    print(f"  - {p.strip()}")

# Look for faithfulness and causality metrics
print("\n\nFaithfulness related:")
faith_matches = re.findall(r'[^.]*faithfulness[^.]*\.', original_doc_text, re.IGNORECASE)
for f in faith_matches[:10]:
    print(f"  - {f.strip()[:200]}")
```
